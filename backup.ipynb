```python
import pandas as pd

import zipfile
from urllib.request import urlopen
import io
```

## Tratamento com pandas arquivo obtido do prismnet (não utilizado, Celestrac já cobre todos)

```python
url = 'https://www.prismnet.com/~mmccants/catalogs/rcs.zip'
zf = zipfile.ZipFile(io.BytesIO(urlopen(url).read()))
zf.filelist
```

```python
colnames=['NORAD_CAT_ID', 'RCS', 'SOURCE'] 
rcs_prismnet = pd.read_fwf(zf.open('rcs'),widths=[5,5,3] ,names=colnames, header=None)
rcs_prismnet
```

```python
rcs_prismnet = rcs_prismnet.drop_duplicates('NORAD_CAT_ID', keep='last')
rcs_prismnet
```

```python
rcs_prismnet = rcs_prismnet[rcs_prismnet.RCS != '0']
rcs_prismnet
```

```python
rcs_prismnet = rcs_prismnet.dropna(subset = ['RCS', 'SOURCE'])
rcs_prismnet
```

```python
rcs_prismnet.to_csv('rcs_prismnet.csv', index=False)
```

## colorir exel

```python
df1 = satcat[['NORAD_CAT_ID', 'RCS']]
df1.reset_index(drop=True, inplace=True)
```

```python
def highlight_cells(val):
    if val < 1.0: color = '#FF4500'
    else:  color = '#1E90FF'
    return 'background-color: {}'.format(color)

df1 = df1.style.applymap(highlight_cells)
with pd.ExcelWriter("path_to_file.xlsx") as writer:
    df1.to_excel(writer, sheet_name='Sheet 1', engine='openpyxl')
```
